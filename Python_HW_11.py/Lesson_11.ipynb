```python
import matplotlib.pyplot as plt
import numpy as np
```

# Построение графика

```python
a0 = 0.6
a1 = 5.5
a2 = 10
a3 = -5

x = np.linspace(-7.5,2.5)
y = a0*x**3 + a1*x**2 + a2*x + a3
plt.grid(True)

ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.plot(x,y)
```

# Нахождение корней

```python
# Решение по формуле Кардано

b1 = a1/a0
b2 = a2/a0
b3 = a3/a0

# map(lambda b: b/a0,[a1,a2,a3])

p = -((b1**2)/3) + b2
q = ((2*(b1**3))/27) - (b1*b2)/3 + b3

y = (-q/2 + ((q**2)/4 + (p**3)/27)**(1/2))**(1/3) + (-q/2 - ((q**2)/4 + (p**3)/27)**(1/2))**(1/3)

x1 = y - b1/3
x2 = y - b2/3
x3 = y - b3/3

# map(lambda x: y - x/3,[b1,b2,b3])

print(x1,x2,x3) # Корни получились непонятные, ниже решение через встроенные функции sympy
```

# Нахождение вершин и интервалов возрастания и убывания
